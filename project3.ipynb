## Explore Categorical Variables

### Subtask:
Analyze the distribution of categorical columns like 'job', 'marital', 'education', and 'default' using value counts and visualize them using bar plots.


**Reasoning**:
To analyze the distribution of categorical variables, I will first import the necessary plotting libraries, then iterate through the specified categorical columns ('job', 'marital', 'education', 'default'), printing their value counts and generating a bar plot for each, ensuring proper labeling and display.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size for better readability of plots
plt.figure(figsize=(10, 6))

# Analyze 'job' column
print("Value counts for 'job' column:")
print(df['job'].value_counts())
sns.countplot(data=df, x='job', palette='viridis')
plt.title('Distribution of Job Categories')
plt.xlabel('Job Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Analyze 'marital' column
print("\nValue counts for 'marital' column:")
print(df['marital'].value_counts())
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='marital', palette='cividis')
plt.title('Distribution of Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Analyze 'education' column
print("\nValue counts for 'education' column:")
print(df['education'].value_counts())
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='education', palette='magma')
plt.title('Distribution of Education Levels')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Analyze 'default' column
print("\nValue counts for 'default' column:")
print(df['default'].value_counts())
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='default', palette='plasma')
plt.title('Distribution of Credit Default Status')
plt.xlabel('Has Default')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

**Reasoning**:
To address the `FutureWarning` regarding the `palette` parameter in `sns.countplot`, I will explicitly assign the `x` variable to `hue` and set `legend=False` in each `countplot` call, as recommended by the warning, to ensure forward compatibility and suppress the warning.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size for better readability of plots

# Analyze 'job' column
print("Value counts for 'job' column:")
print(df['job'].value_counts())
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='job', hue='job', palette='viridis', legend=False)
plt.title('Distribution of Job Categories')
plt.xlabel('Job Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Analyze 'marital' column
print("\nValue counts for 'marital' column:")
print(df['marital'].value_counts())
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='marital', hue='marital', palette='cividis', legend=False)
plt.title('Distribution of Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Analyze 'education' column
print("\nValue counts for 'education' column:")
print(df['education'].value_counts())
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='education', hue='education', palette='magma', legend=False)
plt.title('Distribution of Education Levels')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Analyze 'default' column
print("\nValue counts for 'default' column:")
print(df['default'].value_counts())
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='default', hue='default', palette='plasma', legend=False)
plt.title('Distribution of Credit Default Status')
plt.xlabel('Has Default')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

## Explore Numerical Variables

### Subtask:
Analyze the distribution of numerical columns like 'age' using histograms and box plots with legends.


**Reasoning**:
To analyze the distribution of numerical variables, I will identify the numerical columns, then iterate through each to generate a histogram to visualize its distribution and a box plot to identify potential outliers, ensuring proper labeling and display for each plot.



```python
numerical_cols = ['age', 'duration', 'campaign', 'pdays', 'previous', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']

for col in numerical_cols:
    plt.figure(figsize=(12, 5))

    # Histogram
    plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col.replace("_", " ").title()}')
    plt.xlabel(col.replace("_", " ").title())
    plt.ylabel('Frequency')

    # Box plot
    plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
    sns.boxplot(y=df[col], color='lightcoral')
    plt.title(f'Box Plot of {col.replace("_", " ").title()}')
    plt.ylabel(col.replace("_", " ").title())

    plt.tight_layout()
    plt.show()
```

## Correlation Analysis

### Subtask:
Analyze the correlation between numerical variables present in the dataset and visualize it using a heatmap with legends.


**Reasoning**:
To analyze the correlation between numerical variables, I will calculate the Pearson correlation matrix for the identified numerical columns and visualize it using a seaborn heatmap with annotations for better readability, and then add a title and display the plot.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numerical columns for correlation analysis
numerical_cols = ['age', 'duration', 'campaign', 'pdays', 'previous', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```
